# Export Data:

```python
imex = pd.read_csv('Import_Export.csv')
imex
```

```python
# splitting table on im and ex to get them into two seperate columns by merging later
exports = imex[imex['Flow'] == 'Export']
exports
```

```python
#dropping columns
exports = exports.drop(['Commodity','Weight (kg)', 'Quantity Name', 'Quantity'], axis = 1)
exports
```

```python
imports = imex[imex['Flow'] == 'Import']
imports
```

```python
#dropping columns
imports = imports.drop(['Commodity','Weight (kg)', 'Quantity Name', 'Quantity'], axis = 1)
imports
```

```python
# merging exports and imports into one dataframe
im_ex = pd.merge(left = imports, 
                              right = exports,
                              how = 'left', 
                              left_on = ['Country or Area', 'Year'], 
                              right_on = ['Country or Area', 'Year'])
im_ex
```

```python
# renaming columns
im_ex = im_ex.rename(columns={"Trade (USD)_x" : "import_in_usd", "Trade (USD)_y" : "export_in_usd", "Year": "year", "Country or Area" : "country"})
# dropping flow columns
im_ex = im_ex.drop(['Flow_x','Flow_y'], axis = 1)
im_ex
```

```python
im_ex[im_ex.country.str.contains('Slovenia')]
```

```python
# fill this one NaN with same value as next year
im_ex.export_in_usd = im_ex.export_in_usd.fillna('8315799040')
```

```python
# checking for NaNs
im_ex.isnull().sum()
```

```python
im_ex.year = im_ex.year.astype(int)
im_ex
```

# Fuel production Data:

```python
fuel = pd.read_csv('Fuel_production.csv')
fuel
```

```python
fuel_import = fuel[fuel['Commodity - Transaction'] == 'Fuel oil - Imports']
fuel_import
```

```python
fuel_import = fuel_import.drop(['Commodity - Transaction','Unit', 'Quantity Footnotes'], axis = 1)
fuel_import
```

```python
# renaming columns
fuel_import = fuel_import.rename(columns={"Country or Area" : "country", "Quantity" : "import_in_tonnes", "Year": "year"})
fuel_import
```

```python
# fuel['Commodity - Transaction'].value_counts().head(10)
```

# Merging all DataFrames above into one Big DatFrame:

```python
# merging one by one
tempwheat = pd.merge(left = eu_temp_change, 
                              right = wheat,
                              how = 'left', 
                              left_on = ['country', 'year'], 
                              right_on = ['country', 'year'])
tempwheat
```

```python
tempwheatbarley = pd.merge(left = tempwheat, 
                              right = barley,
                              how = 'left', 
                              left_on = ['country', 'year'], 
                              right_on = ['country', 'year'])
tempwheatbarley
```

```python
tempwheatbarleycattle = pd.merge(left = tempwheatbarley, 
                              right = cattle,
                              how = 'left', 
                              left_on = ['country', 'year'], 
                              right_on = ['country', 'year'])
tempwheatbarleycattle
```

```python
tempwheatbarleycattleforest = pd.merge(left = tempwheatbarleycattle, 
                              right = forest,
                              how = 'left', 
                              left_on = ['country', 'year'], 
                              right_on = ['country', 'year'])
tempwheatbarleycattleforest
```

```python
tempwheatbarleycattleforestinhabitants = pd.merge(left = tempwheatbarleycattleforest, 
                              right = inhabitants,
                              how = 'left', 
                              left_on = ['country', 'year'], 
                              right_on = ['country', 'year'])
tempwheatbarleycattleforestinhabitants
```

```python
tempwheatbarleycattleforestinhabitantsim_ex = pd.merge(left = tempwheatbarleycattleforestinhabitants, 
                              right = im_ex,
                              how = 'left', 
                              left_on = ['country', 'year'], 
                              right_on = ['country', 'year'])
tempwheatbarleycattleforestinhabitantsim_ex
```

```python
tempwheatbarleycattleforestinhabitantsim_ex.isnull().sum()
```

```python
tempwheatbarleycattleforestinhabitantsim_ex[tempwheatbarleycattleforestinhabitantsim_ex.isna().any(axis=1)]
```

```python
# filling NaNs with k neighbor regression
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import MinMaxScaler

# splitting off numericals 
numericals = tempwheatbarleycattleforestinhabitantsim_ex.select_dtypes(np.number)
numericals
```

```python
# selecting numericals without NaNs
numericals_with_import_in_usd = numericals[numericals['import_in_usd'].notna()]
numericals_with_import_in_usd.shape
#X,y Split
X_import_in_usd = numericals_with_import_in_usd.drop(['import_in_usd', 'year', 'export_in_usd' ],axis=1)
y_import_in_usd = numericals_with_import_in_usd['import_in_usd']
# MinMax tranforming
min_max_transformer = MinMaxScaler().fit(X_import_in_usd) 
X_normalized_import_in_usd = min_max_transformer.transform(X_import_in_usd) 
#print(x_normalized.shape)
X_normalized_import_in_usd = pd.DataFrame(X_normalized_import_in_usd, columns=X_import_in_usd.columns)
X_normalized_import_in_usd
# KNN regression
KNN = KNeighborsRegressor(n_neighbors=3)
KNN.fit(X_normalized_import_in_usd,y_import_in_usd)
KNN.score(X_normalized_import_in_usd,y_import_in_usd)
# selecting all numericals
X_all = numericals.drop(['import_in_usd', 'year', 'export_in_usd'],axis=1)
X_normalized_all = pd.DataFrame(MinMaxScaler().fit_transform(X_all), columns=X_all.columns)
y_all = numericals['import_in_usd']
#predicting
predicted_import_in_usd = pd.Series(KNN.predict(X_normalized_all),name='predicted_import_in_usd')
# predicted values into table
tempwheatbarleycattleforestinhabitantsim_ex = pd.concat([tempwheatbarleycattleforestinhabitantsim_ex,predicted_import_in_usd],axis=1)
tempwheatbarleycattleforestinhabitantsim_ex
# filling NaNs with predicted values
tempwheatbarleycattleforestinhabitantsim_ex['import_in_usd_clean'] =  np.where(tempwheatbarleycattleforestinhabitantsim_ex['import_in_usd'].isna(), tempwheatbarleycattleforestinhabitantsim_ex['predicted_import_in_usd'], tempwheatbarleycattleforestinhabitantsim_ex['import_in_usd'])
tempwheatbarleycattleforestinhabitantsim_ex
```

```python
numericals_with_export_in_usd = numericals[numericals['export_in_usd'].notna()]
# numericals_with_export_in_usd.shape

X_export_in_usd = numericals_with_import_in_usd.drop(['import_in_usd', 'year', 'export_in_usd' ],axis=1)
y_export_in_usd = numericals_with_import_in_usd['export_in_usd']

min_max_transformer2 = MinMaxScaler().fit(X_export_in_usd) 
X_normalized_export_in_usd = min_max_transformer.transform(X_export_in_usd) 
#print(x_normalized.shape)
X_normalized_export_in_usd = pd.DataFrame(X_normalized_export_in_usd, columns=X_export_in_usd.columns)
X_normalized_export_in_usd

KNN = KNeighborsRegressor(n_neighbors=3)
KNN.fit(X_normalized_export_in_usd,y_export_in_usd)
display(KNN.score(X_normalized_export_in_usd,y_export_in_usd))

X_all2 = numericals.drop(['import_in_usd', 'year', 'export_in_usd'],axis=1)
X_normalized_all2 = pd.DataFrame(MinMaxScaler().fit_transform(X_all2), columns=X_all2.columns)
y_all2 = numericals['export_in_usd']

predicted_export_in_usd = pd.Series(KNN.predict(X_normalized_all2),name='predicted_export_in_usd')

tempwheatbarleycattleforestinhabitantsim_ex = pd.concat([tempwheatbarleycattleforestinhabitantsim_ex,predicted_export_in_usd],axis=1)
tempwheatbarleycattleforestinhabitantsim_ex

tempwheatbarleycattleforestinhabitantsim_ex['export_in_usd_clean'] =  np.where(tempwheatbarleycattleforestinhabitantsim_ex['export_in_usd'].isna(), tempwheatbarleycattleforestinhabitantsim_ex['predicted_export_in_usd'], tempwheatbarleycattleforestinhabitantsim_ex['export_in_usd'])
tempwheatbarleycattleforestinhabitantsim_ex
```

```python
# dropping predicted and original import and export columns, only keeping the cleaned ones
tempwheatbarleycattleforestinhabitantsim_ex = tempwheatbarleycattleforestinhabitantsim_ex.drop(['import_in_usd', 'export_in_usd', 'predicted_import_in_usd', 'predicted_export_in_usd'], axis = 1)
tempwheatbarleycattleforestinhabitantsim_ex
```
